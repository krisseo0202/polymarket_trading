```python

try:
    from websocket import WebSocketApp
except ImportError:
    print("Warning: websocket-client not installed. WebSocket functionality disabled.")
    WebSocketApp = None
def round_to_tick(price: float, tick_size: float) -> float:
    """Round price to match tick size (guessed if unknown)"""
    if tick_size is None or tick_size <= 0:
        # default tick to 0.001 if unknown
        tick_size = 0.001
    decimals = abs(str(tick_size)[::-1].find('.')) if '.' in str(tick_size) else 0
    return round(round(price / tick_size) * tick_size, decimals)

def get_market_info(token_id: str) -> dict:
    """
    Try to get minimal viable market info (tick_size best guess if missing)
    """
    url = f"https://clob.polymarket.com/book?token_id={token_id}"
    resp = requests.get(url)
    result = resp.json()

    return result

def get_midpoint(token_id: str) -> float:
    """Get current midpoint price or best effort"""
    try:
        mid = client.get_midpoint(token_id)
        if isinstance(mid, dict) and "mid" in mid:
            return float(mid["mid"])
        # fallback to simple orderbook mid
        ob = client.get_order_book(token_id)
        best_ask = float(ob["asks"][0]["p"]) if ob.get("asks") else 0.5
        best_bid = float(ob["bids"][0]["p"]) if ob.get("bids") else 0.5
        return (best_ask + best_bid) / 2.
    except Exception as e:
        raise Exception(f"Failed to get midpoint for token {token_id}: {e}")

def post_two_sided_quotes(
    token_id: str, 
    spread: float = 0.02, 
    size: float = 1
) -> tuple:
    """Post bid and ask orders. Tick size robust even if missing/unknown."""
    try:
        market = get_market_info(token_id)
        tick_size = float(market.get("tick_size", 0.001))
        neg_risk = market.get("neg_risk") or False

        mid = get_midpoint(token_id)
        bid_price = round_to_tick(mid - spread / 2, tick_size)
        ask_price = round_to_tick(mid + spread / 2, tick_size)
        bid_price = max(0.01, min(0.99, bid_price))
        ask_price = max(0.01, min(0.99, ask_price))

        bid = ask = None

        try:
            
            bid = client.create_and_post_order(
                OrderArgs(token_id=token_id, price=bid_price, size=size, side=BUY),
                # options={"tick_size": tick_size, "neg_risk": neg_risk}
            )
        except Exception as e:
            print(f"Warning: Failed to post bid order: {e}")
            bid = None

        try:
            ask = client.create_and_post_order(
                OrderArgs(token_id=token_id, price=ask_price, size=size, side=SELL),
                # options={"tick_size": tick_size, "neg_risk": neg_risk}
            )
        except Exception as e:
            print(f"Warning: Failed to post ask order: {e}")
            if bid:
                try:
                    bid_id = bid.get('id') or bid.get('order_id')
                    if bid_id:
                        client.cancel_order(bid_id)
                except Exception:
                    pass
            ask = None

        return bid, ask, mid

    except Exception as e:
        raise Exception(f"Failed to post two-sided quotes: {e}")

def cancel_all_orders():
    """Cancel all open orders."""
    try:
        client.cancel_all()
    except Exception as e:
        print(f"Warning: Failed to cancel all orders: {e}")

def run_market_maker(token_id: str, spread: float = 0.02, size: float = 100, interval: int = 30):
    """Basic market maker with tick size safety."""
    if not token_id or token_id == "YOUR_TOKEN_ID":
        raise ValueError("Invalid token_id. Please provide a valid token ID.")
    if spread <= 0:
        raise ValueError("spread must be positive")
    if size <= 0:
        raise ValueError("size must be positive")
    if interval <= 0:
        raise ValueError("interval must be positive")
    
    print(f"Starting market maker for token: {token_id}")
    print(f"Spread: {spread}, Size: {size}, Interval: {interval}s")

    while True:
        try:
            cancel_all_orders()
            bid, ask, mid = post_two_sided_quotes(token_id, spread, size)
            bid_status = "✓" if bid else "✗"
            ask_status = "✓" if ask else "✗"
            print(f"Mid: {mid:.4f} | Bid: {bid_status} | Ask: {ask_status}")
            if bid:
                print(f"  Bid order ID: {bid.get('id', bid.get('order_id', 'N/A'))}")
            if ask:
                print(f"  Ask order ID: {ask.get('id', ask.get('order_id', 'N/A'))}")
            time.sleep(interval)

        except KeyboardInterrupt:
            print("\nShutting down...")
            try:
                cancel_all_orders()
            except Exception:
                pass
            break
        except Exception as e:
            print(f"Error: {type(e).__name__}: {e}")
            traceback.print_exc()
            time.sleep(5)

def start_websocket(token_id: str, on_update):
    """Subscribe to real-time orderbook updates."""
    if WebSocketApp is None:
        raise ImportError("websocket-client not installed. Install with: pip install websocket-client")
    url = "wss://ws-subscriptions-clob.polymarket.com/ws/market"

    def on_message(ws, message):
        try:
            data = json.loads(message)
            on_update(data)
        except Exception as e:
            print(f"WebSocket message error: {e}")

    def on_open(ws):
        try:
            ws.send(json.dumps({
                "type": "market",
                "assets_ids": [token_id]
            }))
        except Exception as e:
            print(f"WebSocket open error: {e}")

    def on_error(ws, error):
        print(f"WebSocket error: {error}")

    def on_close(ws, close_status_code, close_msg):
        print("WebSocket connection closed")

    def ping(ws):
        time.sleep(1)
        while True:
            try:
                ws.send("PING")
                time.sleep(10)
            except Exception:
                break

    ws = WebSocketApp(
        url, 
        on_message=on_message, 
        on_open=on_open,
        on_error=on_error,
        on_close=on_close
    )
    threading.Thread(target=ping, args=(ws,), daemon=True).start()
    ws.run_forever()

# Choose a token_id you're interested in
TOKEN_ID = "91737931954079461205792748723730956466398437395923414328893692961489566016241"
run_market_maker(TOKEN_ID, spread=0.02, size=100, interval=30)
```
